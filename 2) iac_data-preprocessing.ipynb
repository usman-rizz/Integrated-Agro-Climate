     <b>Dealing with NAN Values</b>
</h3>

Missing values were handled using a `location-wise`, `time-based interpolation strategy`, which is a standard industry practice for ***meteorological time-series data***..

Instead of applying global statistical measures such as `mean or median`, missing observations were estimated based on adjacent temporal values within the same location. 

This approach `preserves seasonal trends`, `daily variability`, and `physical realism `while ***minimizing information loss***.

* ***Set Date as Time Index***

```python
cleaned_data = cleaned_data.set_index("Date")
```

* ***Sort Data by Location and Date***

```python
cleaned_data = cleaned_data.sort_values(["Location", "Date"])
```

* ***Identify Columns Requiring Imputation***

```python
cols_to_interpolate = [
    
    "Uv_Index",
    "Solar_Radiation_Dni",
    "Specific_Humidity",
    "Relative_Humidity",
    "Rainfall_Mm",
    "Surface_Pressure",
    "Wind_Speed_Max_2m",
    "Wind_Speed_Min_2m",
    "Wind_Speed_Range_2m",
    "Wind_Direction_Deg",
    "Wind_Speed_Min_10m"
]
```

* ***Apply Location-Wise Time-Based Interpolation***

```python
for col in cols_to_interpolate:
    cleaned_data[col] = (

        cleaned_data.groupby("Location")[col].transform(lambda x: x.interpolate(method="time"))
        
    )
```

* ***Handle Boundary Missing Values***

```python
for col in cols_to_interpolate:
    cleaned_data[col] = (
        
        cleaned_data.groupby("Location")[col].transform(lambda x: x.bfill().ffill())

    )
```

* ***Final Missing Value Validation***

```python
cleaned_data.isna().sum()
```

* ***Restore Date as a Column***

```python
cleaned_data = cleaned_data.reset_index()
```
